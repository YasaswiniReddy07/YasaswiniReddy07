Tasks To Be Performed:
1. Data Manipulation:

  ● Extract the 5th column and store it in ‘customer_5’

  ● Extract the 15th column and store it in ‘customer_15’

  ● Extract all the male senior citizens whose payment method is electronic
check and store the result in ‘senior_male_electronic’

  ● Extract all those customers whose tenure is greater than 70 months or
their monthly charges is more than $100 and store the result in
‘customer_total_tenure’

  ● Extract all the customers whose contract is of two years, payment method
is mailed check and the value of churn is ‘Yes’ and store the result in
‘two_mail_yes’

  ● Extract 333 random records from the customer_churndataframe and store
the result in ‘customer_333’

  ● Get the count of different levels from the ‘Churn’ column

```python
import pandas as pd

# Read the CSV data
df = pd.read_csv("customer_churn.csv")

# Extract the 5th and 15th columns
customer_5 = df.iloc[:, 4]  # Access column at index 4 (5th column)
customer_15 = df.iloc[:, 14]  # Access column at index 14 (15th column)

print(customer_5.head())  # Display the first few elements of customer_5
print(customer_15.head())  # Display the first few elements of customer_15
```

```python
# Filter for male senior citizens with electronic check payment
senior_male_electronic = df[(df['gender'] == 'Male')  # Filter by gender
                            & (df['SeniorCitizen'] == 1)  # Filter by age group
                            & (df['PaymentMethod'] == 'Electronic Check')]  # Filter by payment method

print(senior_male_electronic.head())  # Display the first few elements
```

```python
# Filter for long tenure or high monthly charges
customer_total_tenure = df[(df['tenure'] > 70) | (df['MonthlyCharges'] > 100)]

print(customer_total_tenure.head())  # Display the first few elements
```

```python
# Filter for 2-year contracts, mailed check payment, and churned customers
two_mail_yes = df[(df['Contract'] == 'Two years')  # Filter by contract length
                   & (df['PaymentMethod'] == 'Mailed Check')  # Filter by payment method
                   & (df['Churn'] == 'Yes')]  # Filter by churn value

print(two_mail_yes.head())  # Display the first few elements
```

```python
# Sample 333 random rows
customer_333 = df.sample(n=333, random_state=42)  # Set a seed for reproducibility

print(customer_333.head())  # Display the first few elements
```

```python
# Count unique values (levels) in the 'Churn' column
churn_counts = df['Churn'].value_counts()

print(churn_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Bar plot for Internet Service
plt.figure(figsize=(8, 6))  # Set figure size
plt.bar(df['InternetService'].value_counts().index, df['InternetService'].value_counts().values, color='orange')
plt.xlabel('Categories of Internet Service')
plt.ylabel('Count of Categories')
plt.title('Distribution of Internet Service')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram for Tenure
plt.figure(figsize=(8, 6))
plt.hist(df['tenure'], bins=30, color='green')  # 30 bins for histogram
plt.xlabel('Tenure (months)')
plt.ylabel('Frequency')
plt.title('Distribution of Tenure')
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter Plot: Monthly Charges vs. Tenure
plt.figure(figsize=(8, 6))
plt.scatter(df['tenure'], df['MonthlyCharges'], color='brown')
plt.xlabel('Tenure of customer')
plt.ylabel('Monthly Charges of customer')
plt.title('Tenure vs Monthly Charges')
plt.tight_layout()
plt.show()
```

```python
# 4. Box Plot: Tenure vs. Contract
plt.figure(figsize=(8, 6))
df.boxplot(by='Contract', column='tenure', showmeans=True)  # Show means as diamonds
plt.xlabel('Contract')
plt.ylabel('Tenure (months)')
plt.title('Tenure by Contract')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

3. Linear Regression:

  ● Build a simple linear model where dependent variable is ‘MonthlyCharges’
and independent variable is ‘tenure’:
a
  . Divide the dataset into train and test sets in 70:30 ratio.

  b. Build the model on train set and predict the values on test set

  c. After predicting the values, find the root mean square error

  d. Find out the error in prediction & store the result in ‘error’

  e. Find the root mean square error


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = df[['tenure']]
y = df['MonthlyCharges']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
error = y_test - y_pred
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
```

```python
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

4. Logistic Regression:

  ● Build a simple logistic regression model where dependent variable is
‘Churn’ and independent variable is ‘MonthlyCharges’:

  a. Divide the dataset in 65:35 ratio

  b. Build the model on train set and predict the values on test set

  c. Build the confusion matrix and get the accuracy score

  d. Build a multiple logistic regression model where dependent variable
is ‘Churn’ and independent variables are ‘tenure’ and
‘MonthlyCharges’

  e. Divide the dataset in 80:20 ratio

  f. Build the model on train set and predict the values on test set

  g. Build the confusion matrix and get the accuracy score

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
X1 = df[['MonthlyCharges']]  # Independent variable
y1 = df['Churn']  # Dependent variable (binary: Yes/No)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.35, random_state=42)
```

```python
model1 = LogisticRegression()
model1.fit(X1_train, y1_train)
y1_pred = model1.predict(X1_test)
```

```python
cm1 = confusion_matrix(y1_test, y1_pred)
accuracy1 = accuracy_score(y1_test, y1_pred)
```

```python
print("Simple Logistic Regression (Monthly Charges):")
print(f"Confusion Matrix:\n{cm1}")
print(f"Accuracy: {accuracy1:.2f}")
```

```python
X2 = df[['tenure', 'MonthlyCharges']]  # Independent variables
y2 = df['Churn']  # Dependent variable (binary: Yes/No)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
```

```python
model2 = LogisticRegression()
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2_test)
```

```python
cm2 = confusion_matrix(y2_test, y2_pred)
accuracy2 = accuracy_score(y2_test, y2_pred)
```

```python
print("\nMultiple Logistic Regression (Tenure & Monthly Charges):")
print(f"Confusion Matrix:\n{cm2}")
print(f"Accuracy: {accuracy2:.2f}")
```
